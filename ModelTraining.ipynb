```python
# Importing california housing dataset and KNeighborsRegression
from sklearn.datasets import fetch_california_housing
from sklearn.neighbors import KNeighborsRegressor
from matplotlib.pylab import plt
```

```python
# Getting features and values and assign to X,y
X, y = fetch_california_housing(return_X_y=True)
```

```python
# create KNeighborsRegression model
model = KNeighborsRegressor(n_neighbors=3)
```

```python
# Fitting or Training model
model.fit(X,y)
```

```python
# Using trained model to predict
pred = model.predict(X)
```

```python
# Plotting predicted values and actual values
plt.scatter(pred, y)
```
